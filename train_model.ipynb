```python
x_train = train_datagen.flow_from_directory("/home/wsuser/work/New_2/train",target_size = (64,64),batch_size = 32,class_mode = "categorical")
```

```python
x_test = test_datagen.flow_from_directory("/home/wsuser/work/New_2/test",target_size = (64,64),batch_size = 32,class_mode = "categorical")
```

```python
model = Sequential()
```

```python
model.add(Convolution2D(32,(3,3),input_shape = (64,64,3),activation = "relu"))
```

```python
model.add(MaxPooling2D(pool_size = (2,2)))
```

```python
model.add(Flatten())
```

```python
model.add(Dense(units = 128,kernel_initializer = "random_uniform",activation = "relu"))
```

```python
model.add(Dense(units = 128,kernel_initializer = "random_uniform",activation = "relu"))
```

```python
model.add(Dense(units = 24,kernel_initializer = "random_uniform",activation = "softmax"))
```

```python
model.compile(loss = "categorical_crossentropy",optimizer = "adam",metrics = ["accuracy"])
```

```python
model.fit_generator(x_train,steps_per_epoch = 240,epochs = 25,validation_data = x_test,validation_steps = 100)
```

```python
model.save("FashionF4.h5")
```
